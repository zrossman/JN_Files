```python
#Creating a function that checks the board and sees if there is a winner
def win_check(board):
    
    #First we'll check for horizontal wins
    for row in board:
        if row[0] == row[1] and row[0] == row[2] and row[0] == 'X':
            return 'Player one wins!'
        elif row[0] == row[1] and row[0] == row[2] and row[0] == 'O':
            return 'Player two wins!'
        
    
    #Next we'll check for vertical wins
    for i in range(3):
        if board[0][i] == board[1][i] and board[0][i] == board[2][i] and board[0][i] == 'X':
            return 'Player one wins!'
        elif board[0][i] == board[1][i] and board[0][i] == board[2][i] and board[0][i] == 'O':
            return 'Player two wins!'
            
    #Finnaly we check for diagonal wins
    if board[0][0] == board[1][1] and board[0][0] == board[2][2] and board[0][0] == 'X':
        return 'Player one wins!'  
    elif board[0][2] == board[1][1] and board[0][2] == board[2][0] and board[0][2] == 'X':
        return 'Player one wins!'
    elif board[0][0] == board[1][1] and board[0][0] == board[2][2] and board[0][0] == 'O':
        return 'Player two wins!'  
    elif board[0][2] == board[1][1] and board[0][2] == board[2][0] and board[0][2] == 'O':
        return 'Player two wins!'
    
    return 'Tie'
        
```

```python
game = [[0, 0, 0], 
        [0, 0, 0], 
        [0, 0, 0]]


while True:
    
    #Player one makes their move
    while True:
        player_one_move = input('Player one, what is your move? Format row,col: ')
        player_one_move_split = player_one_move.split(',')
        row = (int(player_one_move_split[0])) - 1
        col = (int(player_one_move_split[1])) - 1
        if game[row][col] == 0:
            game[row][col] = 'X'
            for row in game:
                print(row)
            break
        print('''You can't go there''')
        continue
    
    #Checking if player one wins after their move
    result = win_check(game)
    if result == 'Player one wins!':
        print(result)
        break
    
    #Checking game result after game board is full
    num_zeros = 0
    for row in game:
        for i in range(3):
            if row[i] == 0:
                num_zeros += 1         
    if num_zeros == 0:
        result = win_check(game)
        print(result)
        break
    
    #Player two makes their move
    while True:
        player_two_move = input('Player two, what is your move? Format row,col: ')
        player_two_move_split = player_two_move.split(',')
        row = (int(player_two_move_split[0])) - 1
        col = (int(player_two_move_split[1])) - 1
        if game[row][col] == 0:
            game[row][col] = 'O'
            for row in game:
                print(row)
            break
        print('''You can't go there''')
        continue

    #Checking if player two wins after their move
    result = win_check(game)
    if result == 'Player two wins!':
        print(result)
        break
    
    continue
```
